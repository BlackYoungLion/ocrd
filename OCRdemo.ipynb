```python
import easyocr
```

```python
reader = easyocr.Reader(['en','ru'], gpu = False)
```

```python
import PIL
import os
```

```python
###Вывод картиночки###
from PIL import ImageDraw
#im = PIL.Image.open('invoice1.png')
#im = PIL.Image.open('france.jfif')
#im = PIL.Image.open('1.png')
im = PIL.Image.open('eC-dulTAyEQ.jpg')
im
```

```python
###Поиск текста в одной картинке###
bounds = reader.readtext('eC-dulTAyEQ.jpg')
#bounds = reader.readtext('france.jfif')
bounds
```

```python
for i in bounds:
    print(i[1])
```

```python
###отрисовка коробочек в картиночке(имеются баги)###
def  draw_boxes(image,bounds,color='yellow',width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0,*p1,*p2,*p3,*p0], fill=color,width=width)
    return image

draw_boxes(im,bounds)
```

```python
###сортировка файликов в папочке###
unsorted_file_list = os.listdir()
sortetd_file_list = sorted(unsorted_file_list)
print(sortetd_file_list)
```

```python
type(sortetd_file_list)
```

```python
###создание массивика только с картиночками###
image_list = []
for file in sortetd_file_list:
    if '.png' in file:
        image_list.append(file)
    elif '.jpg' in file:
        image_list.append(file)
    elif '.jpeg' in file:
        image_list.append(file)
    elif '.jfif' in file:
        image_list.append(file)
print(image_list)
```

```python
###Прогон картиночек###
pile = []
for image in image_list:
    bond = reader.readtext(image)
    pile.append(bond)
print(pile)
```

```python
###создание массивика строк найденых на картиночках###
ansvers = []
for i in pile:
    for j in i:
        ansvers.append(j[1])
print(ansvers)
        
```
